```python
# Visualization of gift amout per gift allocation

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("advancement_donations_and_giving_demo.xls", engine="xlrd")
gruppen = df.groupby("Gift Allocation")["Gift Amount"].sum()
gruppen.plot(kind="bar", title="Gift Amount per Category")

for index, value in enumerate(gruppen):
    plt.text(index, value, f"{value:,.0f}", ha="center", va="bottom")

plt.xlabel("Category")
plt.ylabel("Gift Amount in €")
```
